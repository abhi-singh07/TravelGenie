```python
import json
import http.client

conn = http.client.HTTPSConnection("booking-com15.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "1f563c76b6mshbd08cde58433274p1d23c7jsn60ab183d374b",
    'x-rapidapi-host': "booking-com15.p.rapidapi.com"
}
arrival_date = '2025-03-22'
departure_date = '2025-03-23'
adults = 2
```

```python
city_name = "austin"
conn.request("GET", f"/api/v1/hotels/searchDestination?query={city_name}", headers=headers)

res = conn.getresponse()
data = res.read()

data_json = json.loads(data.decode("utf-8"))

for dest in data_json['data']:
  if dest['search_type']=="city":
    dest_id = dest['dest_id']
    break

url = f"/api/v1/hotels/searchHotels?dest_id={dest_id}&search_type=CITY&adults={adults}&children_age=10%2C17&room_qty=1&arrival_date={arrival_date}&departure_date={departure_date}&sort_by=bayesian_review_score&currency_code=USD"

conn.request("GET", url, headers=headers)

res = conn.getresponse()
data = res.read()
data_json = json.loads(data.decode("utf-8"))

for hotel in data_json['data']['hotels']:
  hotel_id = hotel['hotel_id']
  conn.request("GET", f"/api/v1/hotels/getDescriptionAndInfo?hotel_id={hotel_id}&languagecode=en-us", headers=headers)
  res = conn.getresponse()
  hotel_details = res.read()
  hotel_details = json.loads(hotel_details.decode("utf-8"))
  hotel['hotel_details'] = hotel_details

print(json.dumps(data_json, indent=4))
```
